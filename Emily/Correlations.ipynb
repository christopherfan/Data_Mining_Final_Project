```python
import pandas as pd
from pandas import read_csv
from mpld3 import plugins
from pandas.tools.plotting import scatter_matrix

features_day= ['tickets_monday', 'tickets_tuesday', 'tickets_wednesdays', 'tickets_thursday', 'tickets_friday', 'tickets_saturday','tickets_sunday']
features_cat=['tickets_am', 'tickets_pm', 'tickets_cat_1','tickets_cat_2','tickets_cat_3','tickets_cat_4','tickets_cat_5','tickets_cat_6']
```

```python
# Best performers
df = read_csv('features_top_performers.csv', sep = ',')

new_df = df[features_day]
```

```python
df.describe()
```

```python
scatter_matrix(new_df, alpha=0.2, figsize=(6, 6), diagonal='kde')
# --> strong correlation for Monday to Friday
```

```python
# Correlation matrix for best performers, features: AM/PM and violation categories
data = df[features_cat]
scatter_matrix(data, alpha=0.2, figsize=(6, 6), diagonal='kde')
```
